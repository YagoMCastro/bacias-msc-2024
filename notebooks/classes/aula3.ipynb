```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 100.0e3, 1001)

K = 2.0 # Condutividade térmica [W/(K.m)]
Q0 = 0.050 # Fluxo térmico superficial [w/m²]
H = 9.6e-10 # Produção de calor radiogênico
T0 = 20
RHO = 2800 # Densidade

t = T0 + (Q0*x)/K - (RHO*H*x**2)/(2*K)
```

```python
plt.plot(t, x/1000)
plt.ylim(100, 0)
```

```python
QM = 25e-3 # Fluxo térmico basal proveniente do manto [W/m²]
HR = 1e4 # Fator de decaimento exponencial com a profundidade [m]
RHOH0 = 2e-6 # Produção de calor radiogênico por metro cúbico [W/m³]
```

```python
t = T0 + QM*x/K +HR**2*RHOH0*(1-np.exp(-x/HR))/K

plt.plot(t, x/1000)
plt.ylim(100, 0)
```
