```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../../data/Densidade radial.txt" , sep = " " , names = ["r", "rho"])
```

```python
plt.plot(df.r, df.rho, ".")
plt.ylabel("Density [kg/m$^3$]")
plt.xlabel("Radius [km]")
```

```python
rr = np.linspace(0, 6371, 600) # km
dd = np.interp(rr, df.r, df.rho)

plt.plot(rr, dd, ".")
```
