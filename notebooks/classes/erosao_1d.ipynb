```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = 0.0  # elevation [m]
u = 500.0 # Uplift [m/Myr]
u /= 1e6 # Uplift [m/yr]

t = 0.0 # time
dt = 1e4 # [yr]
lamb = 2e-7 # [1/yr]

t_vec = t
h_vec = h
u_vec = u
erode_vec = lamb * h
```

```python
while t < 90.0e6:
    h += -lamb*h*dt + u*dt
    t += dt

    t_vec = np.append(t_vec,t)
    h_vec = np.append(h_vec,h)
    u_vec = np.append(u_vec,u)
    erode_vec = np.append(erode_vec,lamb*h)


plt.figure(figsize=(10,5))
plt.plot(t_vec/1e6,h_vec, label="elevation [m]")
plt.plot(t_vec/1e6,erode_vec*1e6, label="erosion rate [m/Myr]")
plt.plot(t_vec/1e6,u_vec*1e6, label="uplift [m/Myr]")
plt.xlabel("Time [Myr]")
plt.legend()
```
