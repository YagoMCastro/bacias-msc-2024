```python
x = np.linspace(-10,10,1001)
dx = x[1] - x[0]

f = np.exp(-x**2)

plt.plot(x,f)
```

```python
f_central = f[1:-1]
f_left = f[:-2]
f_right = f[2:]

second_derivative = (f_left - 2*f_central + f_right)/(dx**2)

plt.plot(x[1:-1],second_derivative)
```
